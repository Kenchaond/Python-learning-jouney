```python
#Generate a random number between 0 and 1
import random
secret = random.random()
print("**Number guessing game**")
guess=float(input("Please enter your guess:"))  #Ask the player to enter a floating-point number
while abs(guess-secret)>0.01:  #When the absolute difference between guessing number and the secret number is bigger than 0.01 loop the following code
  if guess-secret>0.01:  #The condition that guess number greater than the secret number 
    print("Your number is too big")
    guess=float(input("Please enter your guess:",))
  elif abs(guess-secret)>0.01:  #The condition that guess number less than the secret number
    print("Your number is too small")
    guess=float(input("Please enter your guess:",))
#If the the absolute difference between guessing number and the secret number is less than or equal to 0.01 it will not run the while loop       
print("Congratulation!Your guess is close enough.")
print("The secret number is",secret)
```
