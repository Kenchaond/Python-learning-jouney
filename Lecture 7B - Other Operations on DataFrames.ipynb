---
**Example 1:** Finding the number of rows and columns in a DataFrame.

* **df.shape[0]** is the number of rows (i.e. dimension 0)
* **df.shape[1]** is the number of columns (i.e. dimension 1)

```python
# Print the no. of rows and columns in a DataFrame
print("no. of rows:", datadf.shape[0])
print("no. of columns:", datadf.shape[1])
```

---
**Example 2:** Getting the column names from a DataFrame.
* **list(df.columns)** will create a list containing the names of the columns in a DataFrame.



```python
# Getting the column names from a DataFrame
column_names = list(datadf.columns)
print(column_names)
```

---
**Example 3:** If you need to rename the columns of a DataFrame **df**, you can use the **rename()** method. 
* The syntax is **df.rename({*old_name1:new_name1, old_name2:newname2,...*})**.
* The **rename()** method accepts a dictionary as input. In the dictionary, the key is the old column name and the value is the new column name.
* If a specified key does not exist, the method will ignore it.

```python
# Original column names
print(list(datadf.columns))

# Rename Firstname to Fname
# Rename Lastname to Lname
datadf = datadf.rename(columns={"Firstname":"Fname","Lastname":"Lname"})
display(datadf)
```

---
**Example 4:** If you want to set the names of all columns all at once, you can set it using **columns** property of the DataFrame. 
* **df.columns** contains the name of all columns in a data frame. 
* **df.columns=*list_of_column_names*** will rename all column names of a DataFrame using the names provided in the given list. 
* Number of elements in the list should be the same as the number of columns in the DataFrame.


```python
# Original column names
print(list(datadf.columns))

# Rename all column names all at once
datadf.columns=["id","fname","lname","sex","math","eng","chi","gpa","interest","scholarship","LOAN"]
display(datadf)
```
