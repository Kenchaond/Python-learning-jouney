```python
p=0#use to find highest frequency for 1-grams words
l=0#use to find highest frequency for 2-grams words
a=1#stop the loop
b=0#stop the loop
list_onegramHF=[]#list to store highest frequency for 1-grams words
list_twogramHF=[]#list to store highest frequency for 2-grams words
ALL2gramFrequencylist=[]#list to store the frequencies of all 2-grams words
onegramdictionary={}#Dictionary to store the frequencies of all 1-grams words
twogramdictionary={}#Dictionary to store the frequencies of all 2-grams words

while a!=b:
  word=input("Please Enter a sentence with at least 10 words:")#Ask the user to enter a sentence with at least 10 words in it.
  upperword=word.upper()#Convert the sentence into all upper case.
  list1=upperword.split()#Split the sentence into a list
  if len(list1) < 10:#Check if there are at least 10 words in the sentence,Return to enter again
    print("Your sentence is less than 10 words!")
  else:

#Start to find information of one grams words    
    for i in list1:
      if onegramdictionary.get(i) is None:#If the Dictionary contain nothing create a new key with value 1
        onegramdictionary[i]=1
      else:
        onegramdictionary[i]=onegramdictionary[i]+1#If the key is already in the Dictionary then key value + 1
    for s in onegramdictionary:
      if onegramdictionary[s]>p:#Finding the highest frequency for 1-grams words
        p=onegramdictionary[s]
    for g in onegramdictionary.keys():
      if onegramdictionary.get(g)==p:#If the key value in onegramdictionary = highest frequency for 1-grams words append the key to the list_onegramHF
        list_onegramHF.append(g)

#Start to find information of two grams words
    for o in range(len(list1)-1):
      ALL2gramFrequencylist.append(list1[o]+"/"+list1[o+1])#Appending the frequencies of all 2-grams words into ALL2gramFrequencylist
    for j in ALL2gramFrequencylist:
      if twogramdictionary.get(j) is None:#If the Dictionary contain nothing create a new key with value 1
        twogramdictionary[j]=1
      else:
        twogramdictionary[j]=twogramdictionary[j]+1#If the key is already in the Dictionary then key value + 1      
    for t in twogramdictionary:
      if twogramdictionary[t]>l:#Finding the highest frequency for 2-grams words
        l=twogramdictionary[t]
    for h in twogramdictionary.keys():
      if twogramdictionary.get(h)==l:#If the key value in onegramdictionary = highest frequency for 2-grams words append the key to the list_twogramHF
        list_twogramHF.append(h)
#print the frequencies , the highest frequency list and maximum frequency for one-grams words 
    print()    
    print("1-gram frequencies:")
    print(onegramdictionary)
    print()
    print("1-gram with the highest frequency",list_onegramHF)
    print("Maximum frequency is",p)
    print()
#print the frequencies , the highest frequency list and maximum frequency for two-grams words   
    print("2-gram frequencies:")
    print(twogramdictionary)
    print()
    print("2-gram with the highest frequency",list_twogramHF)
    print("Maximum frequency is",l)
    b=1 #Stop the while loop
#There are words and words and words many words Ha Ha Ha Ha
```
