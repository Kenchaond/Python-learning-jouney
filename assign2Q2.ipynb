```python
#Task 1
# import Pandas module
import pandas as pd

# Read XLSX file into DataFrame 
# We are reading "sales" from the file supermarket_sales.xlsx
datadf = pd.read_excel("/content/drive/MyDrive/Data/supermarket_sales.xlsx",sheet_name="sales")
display(datadf.head(10))#display the first 10 rows of the DataFrame.
```

```python
#Task 2
df=pd.DataFrame()#Create a empty DataFrame
a=datadf["Gross Income"].quantile(q=0.75)-datadf["Gross Income"].quantile(q=0.25)#Calculate the interquartile range of Gross Income
b=datadf["Rating"].quantile(q=0.75)-datadf["Rating"].quantile(q=0.25)#Calculate the interquartile range of Rating
c=datadf["Gross Income"].std()/datadf["Gross Income"].mean()#Calculate the coefficient of variation of Gross Income
d=datadf["Rating"].std()/datadf["Rating"].mean()#Calculate the coefficient of variation of Rating
e=datadf["Gross Income"].std()#Calculate the sample standard deviation of Gross Income
f=datadf["Rating"].std()#Calculate the sample standard deviation of Rating
#Create 4 new Variables "Variable","IQR","CV","SD" with their value in the empty DataFrame
df["Variable"]="Gross Income","Rating"
df["IQR"]=a,b
df["CV"]=c,d
df["SD"]=e,f
display(df)#Display the DataFrame stored the statistics

#Task 3
print()#Use some blank lines to separate the output
Total1=datadf.loc[(datadf["Gender"]=="Female")&(datadf["City"]=="Yangon")]#Extract the column that satisfied with the condition from female customers in Yangon
Total2=datadf.loc[(datadf["Gender"]=="Female")&(datadf["City"]=="Naypyitaw")]#Extract the column that satisfied with the condition from female customers in Naypyitaw
print("Total of 'Gross Income' from female customers in Yangon is",Total1["Gross Income"].sum())#print the total Gross Income for female customers in Yangon
print("Total of 'Gross Income' from female customers in Naypyitaw is",Total2["Gross Income"].sum())#print the total Gross Income for female customers in Naypyitaw

#Task 4
print()#Use some blank lines to separate the output
Mode=datadf.loc[(datadf["Gender"]=="Male")]#Extract the data that Gender is Male
Mean=Mode.loc[Mode["Product Line"]=="Health and beauty"]#Extract the data that Product Line is MaleHealth and beauty
print("The most popular product line among male costumers is",Mode["Product Line"].mode()[0])#print the most popular product line among male customers
print("The average quantity per invoice for this product line among male costumers is",Mean[["Quantity"]].mean()[0])#print the average quantity per invoice for the product line found among male customers
```
